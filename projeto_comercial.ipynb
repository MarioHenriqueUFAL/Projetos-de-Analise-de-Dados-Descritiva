#Tratamento dos dados da Tabela "Produtos"

Durante a análise da base de dados, observou-se que produtos idênticos tinham códigos diferentes. Dessa forma, a primeira coisa a ser feita foi modificar os códigos dos produtos, de modo a garantir que produtos idênticos tenham a mesma identificação na base de dados.


Na construção da modelagem do banco de dados, devemos garantir que cada produto seja identificado de maneira única a partir da utilização de uma chave primário. Para a tabela "produtos" a chave primária escolhida é a coluna "ID-Produto", por este motivo é imprescindível garantir que cada um dos produtos tenha uma identificação única, de modo a impedir redundâncias no banco de dados.


A primeira etapa consistiu na determinação de códigos únicos para cada tipo de produto. Em seguida, foram atribuídos códigos únicos para cada tipo de produto de acordo com o fabricante. Pois produtos do mesmo tipo podem ser fabricados por diferentes fabricantes, caracterizando diferença entre os produtos.

Verificando os valores únicos na coluna "categoria"

```python
df['Categoria'].unique()
```

Identificando os diferentes tipos de produtos

```python
df['Produto'].unique()
```

OS novos códigos de produtos deverão seguir a seguinte forma padrão: PROD-001-00

Após "PROD-" os três dígitos identificam o tipo do produto. Os últimos dois dígitos indicam, para o mesmo tipo do produto, o fabricante.

```python
produto = input('Digite o nome do produto: ')
df[df['Produto'] == produto]
```

```python
novo_codigo = input('Insira o novo código: ')
df.loc[df['Produto'] == produto, ['ID-Produto']] = novo_codigo
df[df['Produto'] == produto]
```

Após essa etapa, cada tipo de produto possui uma identificação única

Verificando os códigos associados a cada tipo de produtos, temos:

```python
for i in df['Produto'].unique():
  print('Produto: '+i)
  print(df.loc[df['Produto'] == i,'ID-Produto'].unique())
```

Agora, prosseguimos para associação de um código único para cada tipo de produto e tipo de fabricante.

```python
codigo_produto = input('Digite o código do produto: ')
df[df['ID-Produto'] == codigo_produto]
```

```python
df[df['ID-Produto'] == codigo_produto]['Fabricante'].unique()
```

```python
fabricante = input('Informe o nome do fabricante: ')
df[(df['ID-Produto'] == codigo_produto) & (df['Fabricante'] == fabricante)]
```

```python
novo_codigo = input('Informe o novo código: ')
df.loc[(df['ID-Produto'] == codigo_produto) & (df['Fabricante'] == fabricante), 'ID-Produto'] = novo_codigo
df[(df['ID-Produto'] == novo_codigo) & (df['Fabricante'] == fabricante)]
```

Verificando se para cada tipo de produto o valor da coluna "Categoria" é único.

```python
for i in df['Produto'].unique():
  print('Produto: '+i)
  print(df[df['Produto'] == i]['Categoria'].unique())
```

Interpretando a coluna "Loja"

```python
for i in df['Loja'].unique():
  print('Loja '+i)
  print(df[df['Loja'] == i]['Cidade'].unique())
```

Portanto, cada código de loja corresponde a um estabelecimento específico.

Interpretando a coluna "Vendedor"

```python
for i in df['Vendedor'].unique():
  print('Vendedor '+i)
  print(df[df['Vendedor'] == i]['Loja'].unique())
```

Logo, um vendedor pode estar associado a mais de uma loja.

```python
df.head()
```

Verificando se cada vendedor tem um "ID-Vendedor" único, temos:

```python
for i in df['Vendedor'].unique():
  print('Vendedor '+i)
  print(df[df['Vendedor'] == i]['ID-Vendedor'].unique())
```

Portanto, os vendores tem uma identificação única.

Investigando a coluna "Segmento", temos:

```python
for i in df['Segmento'].unique():
  print('Segmento '+i)
  print(df[df['Segmento'] == i]['Categoria'].unique())
```

Portanto, a coluna "Segmento" é uma característica do cliente.

Investigando a coluna "Custo", temos:

```python
for i in df['Produto'].unique():
  print('Produto '+i)
  print(df[df['Produto'] == i]['Custo'].unique())
```

Então, o custo é uma característica da venda do produto.

Para cada atributo, foi verificado o tamanho máximo para definir o tipo de variável no banco de dados do PostgreSQL

```python
#Verificar o tamanho máximo e mínimo da coluna
# Escolhe a coluna para análise
coluna = 'Fabricante'

# Aplica a função len (comprimento) a cada valor na coluna escolhida
df[coluna] = df[coluna].astype(str) # Converte os valores para string para evitar erros
comprimentos = df[coluna].apply(len)

# Encontra o comprimento mínimo e máximo
minimo = comprimentos.min()
maximo = comprimentos.max()

print(f"Comprimento mínimo na coluna '{coluna}': {minimo}")
print(f"Comprimento máximo na coluna '{coluna}': {maximo}")
```

#Separando os Dados em Tabelas Diferentes

```python
import pandas as pd
```

```python
df = pd.read_excel('Dados_Comerciais_modificado_v2.xlsx')
df.head(10)
```

## Tabela Produto

A tabela produto visa armazenar os dados dos diferentes produtos contidos na base de dados.

```python
df_produto = df[['id_produto', 'produto', 'categoria', 'fabricante']]
```

```python
df_produto.head(20)
```

```python
duplicados = df_produto.duplicated(subset='id_produto', keep='first')
```

```python
if duplicados.any():
    print(f'Foram encontrados {duplicados.sum()} registros duplicados na coluna "id_produto".')
```

```python
df_produto.drop_duplicates(subset='id_produto', keep='first', inplace = True)
```

```python
df_produto.to_excel('table_produto.xlsx', index = False)
```

## Tabela Loja

Separando as informações das lojas da base de dados, temos:

```python
df.head()
```

```python
df_loja = df[['loja','cidade','estado']]
df_loja.head(10)
```

```python
duplicados = df_loja.duplicated(subset='loja', keep='first')

if duplicados.any():
    print(f'Foram encontrados {duplicados.sum()} registros duplicados na coluna "loja".')
else:
    print('Não foram encontrados valores duplicados!')
```

```python
df_loja.drop_duplicates(subset='loja', keep='first', inplace = True)
df_loja.head(20)
```

```python
df_loja['loja'].value_counts()
```

```python
df_loja.to_excel('table_loja.xlsx', index = False)
```

## Tabela Vendedor

```python
df.head()
```

```python
df_vendedor = df[['id_vendedor', 'vendedor']]
df_vendedor.head()
```

```python
duplicados = df_vendedor.duplicated(subset='id_vendedor', keep='first')

if duplicados.any():
    print(f'Foram encontrados {duplicados.sum()} registros duplicados na coluna "id_vendedor".')
else:
    print('Não foram encontrados valores duplicados!')

df_vendedor.drop_duplicates(subset='id_vendedor', keep='first', inplace = True)
df_vendedor.head(20)
```

```python
df_vendedor['id_vendedor'].value_counts()
```
